Now, calculate a measure of the lexical richness of the text. The code shows us that the number of distinct words is just 23% of the total number of words in humor and 12% in romance. The romance and humor are the genres of the brown corpus.

```
def percentage(count, total):
    return 100 * count / total



print("Given: Lexical diversity -- humor=4.3, fiction:romance=8.3")
humor_percentage = percentage(5017,21695)
romance_percentage = percentage(8452,70022)
print("calculated:   Percentage -- humor=%d,  fiction:romance=%d" %(humor_percentage,romance_percentage))
print("The humor is more lexically diverse than romance fiction because its percentage is high (%d) than romance fiction (%d)." %(humor_percentage,romance_percentage))
```

Calculating the occurances of particular length of words in a corpus, with their frequency distributions are...

```
long_words = [w for w in text9 if len(w) == 4]
print("The all four letter words in text5 are:\n", sorted(long_words))
print("The distinct four letter words are",sorted(set(long_words)))
print("The length of all four letter words is %d." %len(sorted(long_words)))
print("The length of all distinct four letter words is %d.",len(sorted(set(long_words))))
print("Four letter word with decreasing order of frequency:\n",FreqDist(long_words).keys())
print("The most common five words with there count and name is:\n", FreqDist(long_words).most_common(5))
print("The frequency of the above words are (as per order):",FreqDist(long_words).freq("that"),",",FreqDist(long_words).freq("with"),",",FreqDist(long_words).freq("Syme"),",",FreqDist(long_words).freq("said"),",",FreqDist(long_words).freq("like") )
```
