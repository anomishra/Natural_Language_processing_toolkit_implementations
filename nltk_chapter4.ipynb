exercise 30. Print the output of your function for SimpleText. Compute compression using the following formula: length of the text resulting from the application of your function (considered as a string) divided by the length of the original text (considered as a string).

To clarify the notion of uniqueness point, consider a text of five words “all apes ate apples”.  By [] we denote the uniqueness point of each word in this text: “ a[l]l ap[e]s a[t]e ap[p]les”. Thus, the output of your  function for this text should be “al ape at app”.
	A trie for this text follows:
	   a  l  l</br> 
     	     p  e  s </br>
	        p  l  e  s </br>
     	     t  e</br>
	
Also,  https://en.wikipedia.org/wiki/Cohort_model
has a paragraph, which provides more verbal description of where a "uniqueness point" stems from. There a term "recognition point" is used for this concept at first. Below is the extract from that paragraph: 
The cohort model consists of three stages: access, selection, and integration.[7] Under this model, auditory lexical retrieval begins with the first one or two speech segments, or phonemes, reach the hearer's ear, at which time the mental lexicon activates every possible word that begins with that speech segment.[8] This occurs during the "access stage" and all of the possible words are known as the cohort.[9] The words that are activated by the speech signal but are not the intended word are often called "competitors."[10]  As more speech segments enter the ear and stimulate more neurons, causing the competitors that no longer match the input to be "kicked out" or to decrease in activation.[8][12] The processes by which words are activated and competitors rejected in the cohort model are frequently called "activation and selection" or "recognition and competition." These processes continue until an instant, called the recognition point,[8] at which only one word remains activated and all competitors have been kicked out. This is also known as the uniqueness point.


```
def exercise30():

    text_test = nltk.word_tokenize(SimpleText)
    def uniqueness_point(word, trie, recognition_point= '', queue= ''):
        if len(word) == 0:
            if len(queue) > 0:
                return queue
            else:
                return recognition_point
        if len(trie[word[0]]) == 1:
            if len(queue) > 0:
                new_buffer_unique = queue
            else:
                new_buffer_unique = recognition_point + word[0]
            return uniqueness_point(word[1:], trie[word[0]], recognition_point + word[0], new_buffer_unique)
        return uniqueness_point(word[1:], trie[word[0]], recognition_point + word[0])

    def insert(trie, key, value):
        if key:
            first, rest = key[0], key[1:]
            if first not in trie:
                trie[first] = {}
            insert(trie[first], rest, value)
        else:
            trie['value'] = value

    def result(text):
        trie = nltk.defaultdict(dict)
        for word in text:
            insert(trie, word, word)
        return [uniqueness_point(w, trie) for w in text]

    def compression(text):
        print(len(result(text)) / len(text))

    output = result(text_test)
    print(output)
    print("Compression:")
    print(compression(text_test))

exercise30()    
```

```
def exercise26():
    # Example of timer usage:
    # print(Timer(lambda: recursive_catalan(n)).timeit(1))

    # as per book - C0 = 1, and Cn+1 = Σ0..n (CiCn-i).
    def recursive_catalan(n):
        if n == 0:
            return 1
        output = 0
        for i in range(n):
            output += recursive_catalan(i) * recursive_catalan(n - i - 1)
        return output


    def dynamic_catalan(n):        # reference: geeksforgeeks.org
        if (n == 0 or n == 1):
            return 1
        number = [0 for i in range(n + 1)]
        number[0] = 1
        number[1] = 1
        for i in range(2, n + 1):
            number[i] = 0
            for j in range(i):
                number[i] = number[i] + number[j] * number[i - j - 1]
        return number[n]


    print("The 10th catalan number: ")
    print("Using recursive function: ",recursive_catalan(10))
    print("Using dynamicProgramming function: ",dynamic_catalan(10))
    print("Time: ", Timer(lambda: recursive_catalan(10)).timeit(1))
    print("The 16th catalan number: ")
    print("Using recursive function: ",recursive_catalan(16))
    print("Using dynamicProgramming function: ",dynamic_catalan(16))
    print("Time: ", Timer(lambda: dynamic_catalan(16)).timeit(1))

    
exercise26()    
```
